```python
import pandas as pd

df = pd.read_csv('2019_kbo_for_kaggle_v2.csv')

df_2018 = df[df['year'] == 2018]
positions = ['포수', '1루수', '2루수', '3루수', '유격수', '좌익수', '중견수', '우익수']

for position in positions:
    position_df = df_2018[df_2018['cp'] == position]
    top_player = position_df.nlargest(1, 'war')[['batter_name','war']]
    print(f'2018년 최고의 WAR 플레이어는 {top_player}\n')
        

condition_df = df[(df['year'] >= 2015) & (df['year'] <= 2018)]

for year in range(2015, 2018):
    print(f'\nTop 10 Players in {year}')
    year_df = condition_df[condition_df['year'] == year]
    top_players = year_df.nlargest(10, 'H')[['batter_name', 'H']]
    top_avg = year_df.nlargest(10, 'avg')[['batter_name', 'avg']]
    top_hr = year_df.nlargest(10, 'HR')[['batter_name', 'HR']]
    top_obp = year_df.nlargest(10, 'OBP')[['batter_name', 'OBP']]
    print(f'Top 10 Players by Hits:\n{top_players}\n')
    print(f'Top 10 Players by Batting Average:\n{top_avg}\n')
    print(f'Top 10 Players by Home Runs:\n{top_hr}\n')
    print(f'Top 10 Players by On-Base Percentage:\n{top_obp}\n')

se_columns = ['R', 'H', 'HR', 'RBI', 'SB', 'war', 'avg', 'OBP', 'SLG', 'salary']
se_df = df[se_columns]

corr = se_df.corr()['salary']
corrval = correlations.drop('salary').idxmax()

print(f'가장 관련이 높은 항목: {corrval}')
print(f'비율: {corr[corrval]}')
```
